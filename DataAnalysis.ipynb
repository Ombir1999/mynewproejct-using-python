```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
encoding = 'utf-8'
import chardet
```

```python
with open('/content/drive/MyDrive/data/covid.csv','rb') as f:
  encoding = chardet.detect(f.read())['encoding']
```

```python
df = pd.read_csv('/content/drive/MyDrive/data/covid.csv',encoding=encoding)
```

```python
df.dropna(inplace=True)
```

```python
df['NewCases'] =df['NewCases'].astype(int)
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Country/Region', data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Population', hue = 'NewCases')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['TotalDeaths'], as_index=False)['TotalRecovered'].sum().sort_values(by='TotalRecovered', ascending=False)
sns.barplot(x = 'TotalDeaths',y= 'TotalRecovered' ,data = sales_age)
```

```python
sales_age = df.groupby(['Country/Region'], as_index=False)['Tests/1M pop'].sum().sort_values(by='Tests/1M pop', ascending=False)
sns.barplot(x = 'Country/Region',y= 'Tests/1M pop' ,data = sales_age)
```

```python
ax = sns.countplot(data = df, x = 'Serious,Critical', hue = 'ActiveCases')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Serious,Critical'], as_index=False)['TotalRecovered'].sum().sort_values(by='TotalRecovered', ascending=False)
sns.barplot(x = 'Serious,Critical',y= 'TotalRecovered' ,data = sales_age)
```
